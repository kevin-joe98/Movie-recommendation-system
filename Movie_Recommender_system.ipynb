```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_dataset = pd.read_csv('movies.csv')
```

```python
movies_dataset.head()
```

```python
movies_dataset.shape
```

```python
selected_features = ['genres', 'tagline',  'keywords',  'overview', 'title', 'cast', 'director' ]
print(selected_features)
```

```python
for feature in selected_features:
    movies_dataset[feature] = movies_dataset[feature].fillna('')
    
```

```python
combined_features = movies_dataset['genres'] +' '+movies_dataset['keywords']+' '+movies_dataset['tagline']+' '+movies_dataset['overview']+' '+movies_dataset['title']
```

```python
print(combined_features)
```

```python
# converting the text data to feature vectors
 
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name = input('Enter your favorite movie name:')
```

```python
list_of_all_titles = movies_dataset['title'].tolist()
```

```python
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
print(close_match)
```

```python
index_of_the_movie = movies_dataset[movies_dataset.title == close_match]['index'].values[0]
```

```python
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
```

```python
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
 
print('Movies suggested for you : \n')
 
i = 1
 
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_dataset[movies_dataset.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

```python
movie_name = input(' Enter your favourite movie name : ')
 
list_of_all_titles = movies_dataset['title'].tolist()
 
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
 
close_match = find_close_match[0]
 
index_of_the_movie = movies_dataset[movies_dataset.title == close_match]['index'].values[0]
 
similarity_score = list(enumerate(similarity[index_of_the_movie]))
 
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
 
print('Movies suggested for you : \n')
 
i = 1
 
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_dataset[movies_dataset.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
